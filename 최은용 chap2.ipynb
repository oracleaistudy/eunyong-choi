```python
#Chapter2
#Supervised train & Unsupervied train & Reinforced train 
#훈련세트(train set)과 테스트세트(test set) 준비
#도미 데이터 준비하기 (도미의 특징을 길이와 무게로 표한한 것, 이런 특징을 특성(feature)이라함:특성은 데이터의 특징)
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

#빙어 데이터 준비하기
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

#도미와 빙어의 데이터 합하기
fish_length = bream_length + smelt_length
fish_weight = bream_weight + smelt_weight

#2차원 데이터로 변환하기
fish_data=[[l,w] for l, w in zip(fish_length, fish_weight)]
fish_target= [1]*35 + [0]*14

print(fish_data)
print(fish_target)

# KNK 알고리즘 import 

from sklearn.neighbors import KNeighborsClassifier
kn= KNeighborsClassifier()

print(fish_data[4])
print(fish_data[0:5]) 
print(fish_data[:5])
print(fish_data[44:])

#훈련 세트로 입력값 중 0부터 34번째 인덱스까지 사용
train_input = fish_data[:35]
#훈련 세트로 타깃값 중 0부터 34번째 인덱스까지 사용
train_target = fish_target[:35]
#테스트 세트로 입력값 중 35번째부터 마지막 인덱스까지 사용
test_input = fish_data[35:]
#테스트 세트로 타깃값 중 35번째부터 마지막 인덱스까지 사용
test_target = fish_target[35:]

kn.fit(train_input, train_target)
kn.score(test_input, test_target)

# 훈련세트에는 도미만, 테스트세트에는 빙어만 있음으로 훈련세트와 테스트세트에 샘플링 편향(sampling bias)이 발생함


#Numpy import
#Numpy는 파이썬의 대표적인 배열(array)라이브러리
#1,2(행x열),3차원(행x열xZ) 배열
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
kn= KNeighborsClassifier()

input_arr = np.array(fish_data)
target_arr = np.array(fish_target)

print(input_arr)
print(target_arr)
print(input_arr.shape)     #array배열 보여주는 코드
print(target_arr.shape)

np.random.seed(42)         #넘파이에서 일정한 결과를 얻으려면 초기에 랜덤시드 지정
index = np.arange(49)      #arange():정수N을 전달하면 0에서 N~1까지 1씩 증가하는 배열 생성 
np.random.shuffle(index)   #shuffle():주어진 배열을 주작위로 섞음
print(index)
print(input_arr[[1,3]])

train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]

print(input_arr[13], train_input[0])

test_input=input_arr[index[35:]]
test_target=target_arr[index[35:]]

#훈련세트와 테스트세트가 잘섞였는지 시각화

import matplotlib.pyplot as plt   
plt.scatter(train_input[:,0], train_input[:,1] )
plt.scatter(test_input[:,0], test_input[:,1])
plt.xlabel('lenght')
plt.ylabel('weigh')
plt.show()

#두 번째 머신러닝 프로그램
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
kn.predict(test_input)
test_target

```

```python
#########################
####2-2 데이터 전처리####
#########################

np.column_stack(([1,2,3],[4,5,6]))  # 리스트를 일렬로 세운 다음 차례대로 나란히 연결, 튜플형
                                    # 리스트 vs 튜플형 차이 : 튜플 수정 불가

fish_data = np.column_stack((fish_length,fish_weight))
print(fish_data[:5])

#넘파이를 활용한 타깃 데이터 생성
print(np.ones(5)) # 값1을 반환 
print(np.zeros(5)) # 값0을 반환

fish_target = np.concatenate((np.ones(35), np.zeros(14)))
print(fish_target)

# 사이킷런으로 훈련세트와 테스트세트 나누기(넘파이 인덱스 직접 섞는 법에서 업그레이드)

from sklearn.model_selection import train_test_split  #model_selection 모듈을 활용
train_input, test_input, train_target, test_target = train_test_split(fish_data,fish_target, random_state=42)
print(train_input.shape,test_input.shape)
print(train_target.shape, test_target.shape)
print(test_target) #도미와 빙어의 비율=2.5:1, but 3.3:1로 나옴 -> 샘플링편향

#stratify 매개변수에 타깃데이터를 전달하면 클래스 비율에 맞게 데이터 나눔.
#위의 변수는 훈련데이터가 작거나 특정 클래스의 샘플 개수가 적을 때 유용
train_input, test_input, train_target, test_target = train_test_split(fish_data,fish_target, stratify=fish_target,random_state=42)
print(test_target) #statify 매개변수로 도미와 빙어의 비율이 2.25:1

############################
#### 수상한 도미 한마리 ####
############################

from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
print(kn.predict([[25,150]]))

plt.scatter(train_input[:,0],train_input[:,1])
plt.scatter(25,150, marker= '^')  # marker는 매개변수는 모양을 지정
plt.xlabel('lenght')
plt.ylabel('weight')
plt.show()

#indexes 배열을 사용해 이웃 샘플 구분
distances, indexes = kn.kneighbors([[25, 150]])
plt.scatter(train_input[:,0],train_input[:,1])
plt.scatter(25,150, marker= '^')  # marker는 매개변수는 모양을 지정
plt.scatter(train_input[indexes,0],train_input[indexes,1], marker='D') #'D'는 마름모표기
plt.xlabel('length')
plt.ylabel('weight')
plt.show()


print(train_input[indexes])  #직접 데이터로 이웃데이터 확인
print(train_target[indexes]) #타겟 데이터로 이웃데이터 확인
#도미보다 빙어 이웃데이터를 더 많이 참고했음
print(distances) #이웃 샘플까지의 거리확인


#x축과 y축의 거리값 맞추기(현재는 x축의 범위가 좁고(1-40) y축의 범위(1-1000)가 넓어서 
#y축으로 조금만 멀어져도 아주 큰 값으로 계산 --> 도미 샘플이 이웃으로 선택되지 못한 이유
#해결: x축의 범위를 0-1000으로 동일하게 맞춰줌
plt.scatter(train_input[:,0],train_input[:,1])
plt.scatter(25,150, marker= '^')  # marker는 매개변수는 모양을 지정
plt.scatter(train_input[indexes,0],train_input[indexes,1], marker='D') #'D'는 마름모표기
plt.xlim(0,1000)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

#데이터전처리(data preprocessing): 데이터 특성을 일정한 기준으로 맞춤
#전처리방법1: 표준점수(standard score, z-score): 
#표준점수 계산법(standardization): 각 특성값이 평균에서 표준편차의 몇 배 만큼 떨어져 있는지 확인

mean= np.mean(train_input, axis=0)
std= np.std(train_input, axis=0)
print(mean, std)
train_scaled=(train_input-mean)/std  #넘파이의 브로드캐스팅(broadcasting)

#################################
####전처리 데이터로 모델 훈련####
#################################

plt.scatter(train_scaled[:,0],train_scaled[:,1])
plt.scatter(25,150,marker='^') # 이 샘플 또한 mean가 std로 계산해줘야함
plt.xlabel('lenght')
plt.ylabel('weight')
plt.show()

new= ([25,150]-mean)/std # new변수를 선언해서 샘플의 mean과 std로 계산
plt.scatter(train_scaled[:,0],train_scaled[:,1])
plt.scatter(new[0], new[1],marker='^') 
plt.xlabel('lenght')
plt.ylabel('weight')
plt.show()

kn.fit(train_scaled, train_target)

#훈련세트와 동일하게 테스트세트도 스케일 변환
test_scaled=(test_input-mean)/std
kn.score(test_scaled,test_target)

print(kn.predict([new])) # = 이상한 도미의 정답 예측 성공


#스케일이 수정된 후 참조이웃 확인
distances, indexes = kn.kneighbors([new])
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0],new[1],marker='^')
plt.scatter(train_scaled[indexes,0],train_scaled[indexes,1], marker ='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
