```python

from tkinter import VERTICAL
from numpy import size
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------------------------------
# 1. CSV  (   )
# -------------------------------------------------------
df_woA      = pd.read_csv("median_accuracy_woA.csv")
df_woB      = pd.read_csv("median_accuracy_woB.csv")
df_woBMI    = pd.read_csv("median_accuracy_woBMI.csv")
df_woD      = pd.read_csv("median_accuracy_woD.csv")
df_woG      = pd.read_csv("median_accuracy_woG.csv")
df_woI      = pd.read_csv("median_accuracy_woI.csv")
df_woP      = pd.read_csv("median_accuracy_woP.csv")
df_woS      = pd.read_csv("median_accuracy_woS.csv")
df_woSandA  = pd.read_csv("median_accuracy_woSandA.csv")

# -------------------------------------------------------
# 2. Accuracy   DataFrame 
# -------------------------------------------------------
models = df_woA["Model"]

accuracy_df = pd.DataFrame({
    "Model": models,
    
    "w/oAge":     df_woA["Accuracy"],
    "w/oBloodPressure":     df_woB["Accuracy"],
    "w/oBMI":   df_woBMI["Accuracy"],
    "w/oDPF":     df_woD["Accuracy"],
    "w/oGlucose":     df_woG["Accuracy"],
    "w/oInsulin":     df_woI["Accuracy"],
    "w/oPregnancies":     df_woP["Accuracy"],
    "w/oSkinThickness":     df_woS["Accuracy"],
    "w/oSkinandAge": df_woSandA["Accuracy"]
})
accuracy_df.set_index("Model", inplace=True)

# -------------------------------------------------------
# 3.  Accuracy  
#   x: Original / Mean / Median
#   y: Accuracy 
# -------------------------------------------------------
plt.figure(figsize=(10, 6))

for model in accuracy_df.index:
    plt.plot(
        ["w/oAge", "w/oBloodPressure", "w/oBMI", "w/oDPF", "w/oGlucose", "w/oInsulin", "w/oPregnancies", "w/oSkinThickness",  "w/oSkinandAge" ],
        accuracy_df.loc[model],
        marker='o',
        label=model
    )

#plt.xlabel("Data Type", fontsize = 15)
plt.xticks(fontsize= 15, rotation =45, ha = 'right')
plt.yticks(fontsize=15)
plt.ylabel("Accuracy", fontsize = 15)
plt.title("Accuracy Comparison Across Data Set Selection", fontsize=20)
plt.legend()
plt.grid(False)
plt.tight_layout()
plt.show()
```
