```python
# 0이 하나라도 있는 컬럼 확인

df = pd.read_csv('diabetes.csv')

zero_exist = (df == 0).any()
print(zero_exist)
```

```python
# Pregnancies를 제외한 나머지 컬럼에서, 값이 0인 것을 median으로 치환

cols = df.columns[df.columns != "Pregnancies"]

for col in cols:
    median_value = df[col].median()
    df.loc[df[col] == 0, col] = median_value

print(df.head())

# 파일저장
df.to_csv("diabetes_median.csv", index=False)
```

```python
import pandas as pd

# 1. 원본 데이터 다시 불러오기
df = pd.read_csv("diabetes.csv")

# 2. 제외할 컬럼 설정
exclude_cols = ['Pregnancies', 'Outcome']

# 3. 나머지 컬럼만 선택
cols = [col for col in df.columns if col not in exclude_cols]

# 4. 0을 각 컬럼의 평균으로 치환
for col in cols:
    mean_value = df[col].mean()
    df.loc[df[col] == 0, col] = mean_value

# 5. 확인용: 각 컬럼에 0이 남아 있는지 체크
print((df[cols] == 0).sum())   # 모두 0이어야 정상
print(df['Outcome'].value_counts())  # 0,1 개수 확인

# 6. 파일 저장
df.to_csv("diabetes_mean_clean.csv", index=False)
```
